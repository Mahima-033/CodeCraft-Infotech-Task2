```python
#Importing necessary libraries
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Using provided dataset (titanic dataset) - train set
```

```python
df = pd.read_csv("train (1).csv")

df.head()
```

```python

df.info()
```

```python

df.describe()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
```

```python
df.drop(columns="Cabin",inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
## using countplot for survived -

## survival count
sns.countplot(data=df,x="Survived",color="r")
plt.title("Survivors(0=No , 1=Yes)")
plt.show()
```

```python
## survivors by Gender

sns.countplot(data=df,x="Survived",hue="Sex",color="pink")
plt.title("Survivors by Gender")
plt.show()
```

```python

## survival by class

sns.countplot(data=df,x="Survived",hue="Pclass",color="blue")
plt.title("Survival by Class")
plt.show()
```

```python

#  Age Distribution graph -

plt.hist(df['Age'], bins=20, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#  Correlation Heatmap -

numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Boxplot of Fare by Passenger Class (with Outliers)-

plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=df,color="orange")
plt.title('Fare Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
